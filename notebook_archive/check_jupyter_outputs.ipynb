```python
import os
import json
import glob

def has_outputs(notebook_path):
    with open(notebook_path, 'r') as f:
        nb = json.load(f)
    
    for cell in nb['cells']:
        if cell.get('outputs') or cell.get('execution_count'):
            return True
    return False
```

```python
all_jupyter_notebooks = glob.glob('*.ipynb')
all_jupyter_notebooks 
```

```python
# Check all notebooks
for nb in all_jupyter_notebooks:
    if has_outputs(nb):
        print(f"{nb} has outputs")
```
