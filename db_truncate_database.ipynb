```python
import psycopg2
from dotenv import load_dotenv
import os
```

### Cave! This is not reversible!
### Make sure you connect to the correct database!

```python
# To connect to the correct database set
# the correct env and the correct database name below:

database_name = 'image_analysis_dev'
#database_name = 'image_analysis_test'

db_env = '.env'
#db_env = '.env.test'

# Load test environment (override=True ensures it replaces any previously loaded values)
load_dotenv(db_env, override=True)

# Explicit check — you see exactly what you're connecting to
db_name = os.getenv('DB_NAME')
print(f"Connecting to: {db_name}")
assert db_name == database_name, f"Wrong database loaded: {db_name}"


print("✅ Libraries loaded")
```

```python


conn = psycopg2.connect(
    dbname=os.getenv('DB_NAME'),
    user=os.getenv('DB_USER'),
    password=os.getenv('DB_PASSWORD', ''),
    host=os.getenv('DB_HOST'),
    port=os.getenv('DB_PORT')
)
cur = conn.cursor()

# Delete in order (respects foreign keys)
tables = [
    'clustering_results',
    'llm_responses', 
    'predictions',
    'ground_truth_history',
    'analysis_runs',
    'prompts',
    'images'
]

# for table in tables:
#     cur.execute(f"DELETE FROM {table}")
#     print(f"Deleted from {table}")

cur.execute("TRUNCATE images, ground_truth_history, analysis_runs, predictions, llm_responses, clustering_results, prompts RESTART IDENTITY CASCADE")
conn.commit()

conn.commit()
cur.close()
conn.close()
print("✅ All data deleted")
```
